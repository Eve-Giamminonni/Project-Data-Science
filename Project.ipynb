```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "California_Fire_Incidents.csv"
df = pd.read_csv(file_path)

# Display basic information about the dataset
df.info()
print("\nFirst few rows of the dataset:")
print(df.head())

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values:")
print(missing_values[missing_values > 0])

# Convert date columns to datetime format
df['Started'] = pd.to_datetime(df['Started'], errors='coerce')
df['Updated'] = pd.to_datetime(df['Updated'], errors='coerce')

# Summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())

# Visualizing the distribution of acres burned
plt.figure(figsize=(10, 5))
sns.histplot(df['AcresBurned'].dropna(), bins=50, kde=True)
plt.title('Distribution of Acres Burned')
plt.xlabel('Acres Burned')
plt.ylabel('Frequency')
plt.show()

# Count of incidents per county
top_counties = df['Counties'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_counties.index, y=top_counties.values)
plt.xticks(rotation=45)
plt.title('Top 10 Counties with Most Fire Incidents')
plt.xlabel('County')
plt.ylabel('Number of Incidents')
plt.show()
```
